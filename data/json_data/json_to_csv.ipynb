```python
import pandas as pd
```

```python
cat = ['new', 'dresses', 'outwear', 'tops', 'bottoms', 'skirts', 'sale']
df = []
for i in range(len(cat)):
    data = pd.read_json(f'./item_{cat[i]}.json')
    data['category'] = cat[i]
    df.append(data)
item_info = pd.concat(df, ignore_index=True)
item_info.info()
```

```python
item_info.sample(14)
```

```python
name_dup = item_info.loc[item_info.item_name.duplicated()]
item_info.drop(name_dup.index, axis=0, inplace=True)
item_info.reset_index(drop=True, inplace=True)
item_info.info()
```
